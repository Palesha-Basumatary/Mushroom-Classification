## Modelling Notebook

This notebook is to be used for testing out the various models that you want to use. No preprocessing will be done in this notebook. Steps:

1. Read in `data/final_data.csv` that you created in the `Data Cleaning.ipynb`
2. Try various models and print appropriate metrics (accuracy/MSE etc)
3. Pick a final model and save it as `models/model.pkl`

```python
import numpy as np
import pandas as pd
```

```python
data =pd.read_csv("../data/final_dataset.csv")
data.head() 
```

```python
y = data["class"]
x=data.copy()
x.drop("class",axis=1,inplace=True)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, random_state = 0)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,mean_squared_error,precision_score
y_pred1=model1.predict(x_test)
print("Accuracy for Logistic Regression is ",accuracy_score(y_test,y_pred1))
print("Mean Squared Error for Logistic Regression is ",mean_squared_error(y_test,y_pred1))
print("Precision for Logistic Regression is ",precision_score(y_test, y_pred1))
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model2=GaussianNB()
model2.fit(x_train,y_train)
```

```python
y_pred2=model2.predict(x_test)
print("Accuracy for Naive Bayes is ",accuracy_score(y_test,y_pred2))
print("Mean Squared Error for Naive Bayes is ",mean_squared_error(y_test,y_pred2))
print("Precision for Naive Bayes is ",precision_score(y_test, y_pred2))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier()
fitted  = model3.fit(x_train,y_train)
```

```python
y_pred3=model3.predict(x_test)
print("Accuracy for Naive Bayes is ",accuracy_score(y_test,y_pred3))
print("Mean Squared Error for Naive Bayes is ",mean_squared_error(y_test,y_pred3))
print("Precision for Naive Bayes is ",precision_score(y_test, y_pred3))
```

```python
import pickle
filepath = r'../models/Model.pkl'
pickle.dump(model3, open(filepath, 'wb'))
```
